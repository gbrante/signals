### 3. Exemplos

#### 3.1 Vamos usar o NumPy para traçar um seno discreto no tempo:
$$\Large x[n] = \sin\left(\omega n + \phi\right)$$

onde $\large n$ é um vetor de índices sobre o qual a função deve ser avaliada, $\large \omega$ especifica a frequência angular e $\large \phi$ a fase da senóide.

Observe o trecho de código abaixo e procure identificar as semelhanças/diferenças em relação a um código em MATLAB.

```python
from numpy import *
from matplotlib.pyplot import *

n = arange(0,31)
ya = sin(pi/8*n)
yb = 5*sin(pi/8*n + pi/4)
yc = cos(pi/8*n)

figure(1)
stem(n, ya)
axis([0, 30, -1, 1])
grid(True)
xlabel('n', fontsize=18);
ylabel('Amplitude', fontsize=18);
title('(a)');

figure(2)
stem(n, yb)
axis([0, 30, -5, 5])
grid(True)
xlabel('n', fontsize=18);
ylabel('Amplitude', fontsize=18);
title('(b)');

figure(3)
stem(n, yc)
axis([0, 30, -1, 1])
grid(True)
xlabel('n', fontsize=18);
ylabel('Amplitude', fontsize=18);
title('(c)');
```

#### 3.2 Exponencial Complexa Discreta no Tempo:
$$\Large x[n] = 3 \, e^{j \, \frac{\pi}{10} \, n}$$

Observe o trecho de código abaixo e procure identificar as semelhanças/diferenças em relação a um código em MATLAB.

```python
n = arange(0,31)
x = 3*exp(pi/10*1j*n)

figure(1)
f, (ax1, ax2) = subplots(2, 1)
ax1.stem(n, x.real)
ax1.set_ylabel('Parte Real', fontsize=18);
ax2.stem(n, x.imag)
ax2.set_ylabel('Parte Imag.', fontsize=18);
ax2.set_xlabel('n', fontsize=18);

figure(2)
f, (ax1, ax2) = subplots(2, 1)
ax1.stem(n, abs(x))
ax1.set_ylabel('Magnitude', fontsize=18);
ax2.stem(n, angle(x))
ax2.set_ylabel('Fase', fontsize=18);
ax2.set_xlabel('n', fontsize=18);
```

#### 3.3 Função para gerar um _impulso discreto no tempo_:

\begin{equation}
\Large
    \delta[n] = 
    \begin{cases}
            0, \qquad n \neq 0 \\
            1, \qquad n = 0
    \end{cases}
\end{equation}

O trecho de código a seguir cria a funçao <font color=blue>impulso</font>(n) e plota o impulso deslocado $\delta[n-5]$ no intervalo de tempo $-10 \leq n \leq 10$.

Ao final, experimente remover o comentário de **%matplotlib qt** e observe o resultado.

```python
from numpy import *
from matplotlib.pyplot import *
#%matplotlib qt

def impulso(n):
    d = 1*(n==0)
    return d

n = arange(-10,11)
y = impulso(n-5)

stem(n, y)
xlabel('n', fontsize=18);
ylabel('Amplitude', fontsize=18);
title('Função Impulso Atrasada no Tempo');
```
