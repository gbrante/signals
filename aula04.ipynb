### Exemplo 4.4:
Encontrar e traçar a Transformada de Fourier de:

$\Large x(t) = e^{-at} u(t), \qquad a>0$

```python
from numpy import *
from matplotlib.pyplot import *

a = 0.5
omega = arange(-2*pi, 2*pi, 0.001)

X = 1/(a + 1j*omega)

figure(figsize=(7, 5))
plot(omega,abs(X))
axis([-2*pi, 2*pi, 0, 2])
grid(True)
xlabel('$\omega$',fontsize=16)
ylabel('Magnitude',fontsize=16)

figure(figsize=(7, 5))
plot(omega,angle(X))
axis([-2*pi, 2*pi, -2, 2])
grid(True)
xlabel('$\omega$',fontsize=16)
ylabel('Fase',fontsize=16);
```

### Exemplo 4.5:
Traçar a Transformada de Fourier de:

$\Large x[n] = a^n u[n], \qquad |a| < 1$

```python
from numpy import *
from matplotlib.pyplot import *

omega = arange(-2*pi, 2*pi, 0.001)

a = 0.5;
X1 = 1/(1 - a*exp(-1j*omega))

a = -0.5;
X2 = 1./(1 - a*exp(-1j*omega));

figure(figsize=(14, 10))
subplot(2,2,1)
plot(omega,abs(X1))
axis([-2*pi, 2*pi, 0, 2])
grid(True)
ylabel('Magnitude',fontsize=16)
title('a = 0.5',fontsize=16)

subplot(2,2,3)
plot(omega,angle(X1))
axis([-2*pi, 2*pi, -2, 2])
grid(True)
ylabel('Fase',fontsize=16)
xlabel('$\omega$',fontsize=16)

subplot(2,2,2)
plot(omega,abs(X2))
axis([-2*pi, 2*pi, 0, 2])
grid(True)
title('a = -0.5',fontsize=16)

subplot(2,2,4)
plot(omega,angle(X2))
axis([-2*pi, 2*pi, -2, 2])
grid(True)
xlabel('$\omega$',fontsize=16);
```
