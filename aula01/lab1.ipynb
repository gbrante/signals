Para essa atividade, vamos escrever funções em Python para criar alguns sinais básicos e usar essas funções para plotar esses sinais no tempo discreto. Por exemplo, vamos escrever uma função para criar um _impulso discreto no tempo_:

\begin{equation}
    \delta[n] = 
    \begin{cases}
            0, \qquad n \neq 0 \\
            1, \qquad n = 0
    \end{cases}
\end{equation}

O trecho de código a seguir cria a funçao <font color=blue>impulso</font>(n) e plota o impulso deslocado $\delta[n-5]$ no intervalo de tempo $-10 \leq n \leq 10$:

```python
import numpy as np
import matplotlib.pyplot as plt

def impulso(n):
    d = 1*(n==0)
    return d

n = np.array(range(-10,11))
y = impulso(n-5)

plt.stem(n, y)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('Função Impulso Atrasada no Tempo');
```

**1.1) Funções Impulso:** Utilizando a função <font color=blue>impulso</font> recém criada, plote as seguintes sequências nos intervalos indicados:

**(a)** $x[n] = 1.5\, \delta[n+3], \qquad\qquad\qquad\qquad\quad -5 \leq n \leq 5$


**(b)** $x[n] = 2.5\, \delta[n+20] - 0.5\,\delta[n-10], \qquad -30 \leq n \leq 30$

```python
na = np.arange(-5,6)
ya = 1.5*impulso(na+3)

nb = np.arange(-30,31)
yb = 2.5*impulso(nb+20) - 0.5*impulso(nb-10)

plt.subplot(1, 2, 1)
plt.stem(na, ya)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('1.1(a)');

plt.subplot(1, 2, 2)
plt.stem(nb, yb)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('1.1(b)');

plt.subplots_adjust(top=1, right=2, wspace=0.5)
```

**1.2)** Vamos agora escrever uma função para gerar o _degrau unitário_:

\begin{equation}
    u[n] = 
    \begin{cases}
            0, \qquad n < 0 \\
            1, \qquad n \geq 0
    \end{cases}
\end{equation}

O trecho de código a seguir cria a funçao <font color=blue>degrau</font>(n) e plota o degrau unitário no intervalo de tempo $-10 \leq n \leq 10$:

```python
def degrau(n):
    d = 1*(n>=0)
    return d

n = np.arange(-10,11)
y = degrau(n)

plt.stem(n, y)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('Função Degrau');
plt.xticks(np.arange(-10,11,2));
```

Use essa função para plotar as seguintes sequências sobre os intervalos indicados:

**(a)** $x[n] = 3.5\, u[n-3], \qquad\qquad\quad -10 \leq n \leq 10$

**(b)** $x[n] = u[n+4] - u[n-4], \qquad -10 \leq n \leq 10$

```python
n = np.arange(-10,11)
ya = 3.5*degrau(n-3)

yb = degrau(n+4) - degrau(n-4)

plt.subplot(1, 2, 1)
plt.stem(n, ya)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('1.2(a)');
plt.xticks(np.arange(-10,11,2));

plt.subplot(1, 2, 2)
plt.stem(n, yb)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('1.2(b)');
plt.xticks(np.arange(-10,11,2));

plt.subplots_adjust(top=1, right=2, wspace=0.5)
```

**1.3)** Vamos agora escrever uma função para gerar uma _função retângulo_:

\begin{equation}
    \mathrm{retangulo}_N[n] = 
    \begin{cases}
            1, \qquad -N \leq n \leq N \\
            0, \qquad |n| > N
    \end{cases}
\end{equation}

O trecho de código a seguir cria a funçao <font color=blue>retangulo</font>(n,N) e plota o retângulo de amplitude unitária no intervalo de tempo $-10 \leq n \leq 10$:

```python
def retangulo(n,N):
    d = 1*(np.abs(n) <= N)
    return d

n = np.arange(-10,11)
y = retangulo(n,4)

plt.stem(n, y)
plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('Função Retângulo');
plt.xticks(np.arange(-10,11,2));
```

Use essa função para plotar as seguintes sequências sobre os intervalos indicados:

**(a)** $x[n] = 5\,\mathrm{retangulo}_4[n-3], \qquad\quad\quad -10 \leq n \leq 10$

**(b)** $x[n] = \mathrm{retangulo}_{10}[n] - \mathrm{retangulo}_5[n], \qquad -20 \leq n \leq 20$

```python
na = np.arange(-10,11)
ya = 5*retangulo(na-3, 4)

nb = np.arange(-20,21)
yb = retangulo(nb, 10) - retangulo(nb, 5)

plt.subplot(1, 2, 1)
plt.stem(na, ya)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('1.3(a)');
plt.xticks(np.arange(-10,11,2));

plt.subplot(1, 2, 2)
plt.stem(nb, yb)
#plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('1.3(b)');
plt.xticks(np.arange(-20,21,2));

plt.subplots_adjust(top=1, right=2, wspace=0.5)
```

**Senóide Discreta no Tempo:** Vamos usar o Python para traçar um seno discreto no tempo:
$$x[n] = \sin\left(\omega n + \phi\right)$$

onde `n` é um vetor de índices sobre o qual a função deve ser avaliada, $\omega$ especifica a frequência angular e $\phi$ a fase da senóide.

Use essa função para plotar (usando a função <font color=blue>stem</font>) as seguintes sequências sobre os intervalos indicados:

**(a)** $x[n] = \sin\left(\frac{\pi}{20}\, n\right), \qquad\qquad\quad 0 \leq n \leq 60$

**(b)** $x[n] = 5\sin\left(\frac{\pi}{10}\, n + \frac{\pi}{4}\right), \qquad -10 \leq n \leq 30$

**(c)** $x[n] = \cos\left(2\pi\frac{1}{5\sqrt{2}}\, n\right), \qquad\quad 0 \leq n \leq 30$

Determine analiticamente se cada sequência é periódica ou não e, em caso afirmativo, determine seu período. Seus plots confirmam essas previsões?

```python
n = np.arange(-10,61)
ya = np.sin(np.pi/20*n)
yb = np.sin(np.pi/10*n + np.pi/4)
yc = np.sin(2*np.pi/(5*np.sqrt(2))*n)

plt.figure(1)
plt.stem(n, ya)
plt.axis([0, 60, -1, 1])
plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('(a)');

plt.figure(2)
plt.stem(n, yb)
plt.axis([-10, 30, -1, 1])
plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('(b)');

plt.figure(3)
plt.stem(n, yc)
plt.axis([0, 30, -1, 1])
plt.grid(True)
plt.xlabel('n', fontsize=18);
plt.ylabel('Amplitude', fontsize=18);
plt.title('(c)');
```

**Exponencial Complexa Discreta no Tempo:** Vamos usar o Python para gerar a seguinte exponencial complexa discreta no tempo:
$$x[n] = 3 \, e^{j \, \frac{\pi}{10} \, n}\qquad\qquad\quad 0 \leq n \leq 30$$

**(a)** Usando os comandos <font color=blue>real</font> e <font color=blue>imag</font>, plote as partes real e imaginária desta função. Use <font color=blue>subplot</font> para obter ambos os plots na mesma figura.

**(b)** Usando os comandos <font color=blue>abs</font> e <font color=blue>angle</font>, plote a magnitude e a fase desta função. Use <font color=blue>subplot</font> para obter ambos os plots na mesma figura.

**(c)** Obtenha as expressões analíticas das partes real e imaginária, bem como magnitude e fase, e compare essas expressões com suas curvas dos itens (a) e (b).

```python
n = np.arange(0,31)
x = 3*np.exp(np.pi/10*1j*n)

plt.figure(1)
plt.subplot(1, 2, 1)
plt.stem(n, x.real)
plt.xlabel('n', fontsize=18);
plt.ylabel('Real', fontsize=18);

plt.subplot(1, 2, 2)
plt.stem(n, x.imag)
plt.xlabel('n', fontsize=18);
plt.ylabel('Imaginário', fontsize=18);

plt.subplots_adjust(top=1, right=2, wspace=0.5)


plt.figure(2)
plt.subplot(1, 2, 1)
plt.stem(n, np.abs(x))
plt.xlabel('n', fontsize=18);
plt.ylabel('Magnitude', fontsize=18);

plt.subplot(1, 2, 2)
plt.stem(n, np.angle(x))
plt.xlabel('n', fontsize=18);
plt.ylabel('Fase', fontsize=18);

plt.subplots_adjust(top=1, right=2, wspace=0.5)
```
